```python
### Accessing List Elements
fruits = ["apple", "banana", "cherry", "date", "Kiwi"]
print(fruits[0])  # Output: apple
print(fruits[2])  # Output: cherry
print(fruits[-1]) # Output: Kiwi
```

```python
print(fruits[1:])  # Output: ['banana', 'cherry', 'date']
print(fruits[:3])  # Output: ['apple', 'banana', 'cherry']
print(fruits[1:4]) # Output: ['banana', 'cherry', 'date']
print(fruits[-3:]) # Output: ['cherry', 'date', 'Kiwi']
print(fruits[:-2]) # Output: ['apple', 'banana', 'cherry']
print(fruits[::2]) # Output: ['apple', 'cherry', 'Kiwi'], True, None]
print(fruits[-1:-3])
```

```python
## Modifying Lists
fruits[1] = "blueberry"
print(fruits)  # Output: ['apple', 'blueberry', 'cherry',
```

```python
fruits = ["apple", "banana", "cherry", "date", "Kiwi"]
fruits.append("orange")
print(fruits)  # Output: ['apple', 'banana', 'cherry', 'date', 'Kiwi', 'orange']
```

```python
## List Methods
fruits.append("orange")  ## Adding an Element
print(fruits)  # Output: ['apple', 'banana', 'cherry', 'date', 'Kiwi', 'orange']
```

```python
fruits.insert(1, "blueberry")  ## Inserting an Element at Specific Position
print(fruits)  # Output: ['apple', 'blueberry', 'banana', '
```

```python
fruits.remove("orange")  ## Removing the first Occurrence of an Element
print(fruits)  # Output: ['apple', 'blueberry', 'banana', '
```

```python
## Remove and Return the last Element
last_fruit = fruits.pop()
print(last_fruit)  # Output: orange
print(fruits)      # Output: ['apple', 'blueberry', 'banana', '
```

```python
index = fruits.index("cherry")
print(index)  
```
