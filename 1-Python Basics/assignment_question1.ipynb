**Question 4:** Write a Python program to find the largest of three numbers.

```python
## Write a Python program to find the largest of three numbers.
user_input1 = input("Enter first number: ")
user_input2 = input("Enter second number: ")
user_input3 = input("Enter third number: ")
num1 = float(user_input1)
num2 = float(user_input2)
num3 = float(user_input3)
if (num1 >= num2) and (num1 >= num3):
    largest = num1
elif (num2 >= num1) and (num2 >= num3):
    largest = num2
else:
    largest = num3
print("The largest number is:", largest)
```

**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Write a Python program to calculate the factorial of a number.
user_input = input("Enter a non-negative integer: ")
num = int(user_input)
factorial = 1
if num < 0:
    print("Factorial is not defined for negative numbers.")
elif num == 0:
    print("The factorial of 0 is 1.")
else:
    for i in range(1, num + 1):
        factorial *= i
    print("The factorial of", num, "is", factorial)
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python
# Write a Python program to check if a year is a leap year.
year = int(input("Enter a year: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")


# Alternative using calendar module
import calendar

year = int(input("Enter a year: "))
if calendar.isleap(year):
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

**Question 19:** Write a Python program to check if a string is a palindrome.

```python
input_str = input("Enter a string: ")
reversed_str = input_str[::-1]
if input_str == reversed_str:
    print(input_str, "is a palindrome.")
else:
    print(input_str, "is not a palindrome.")
```
