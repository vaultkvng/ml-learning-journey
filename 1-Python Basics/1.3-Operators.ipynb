```python
## Simple Calculator
num1=float(input("Enter first number: "))
num2=float(input("Enter second number: "))
operation=input("Enter operation (+, -, *, /, %, e): ")
if operation=='+':
    print(f"{num1} + {num2} = {num1+num2}")
elif operation=='-':
    print(f"{num1} - {num2} = {num1 - num2}")
elif operation=='*':
    print(f"{num1} * {num2} = {num1 * num2}")
elif operation=='/':
    print(f"{num1} / {num2} = {num1 / num2}")
elif operation=='%':
    print(f"{num1} % {num2} = {num1 % num2}")
elif operation=='e':
    print(f"{num1} ** {num2} = {num1 ** num2}")
else:
    print("Invalid operation")
    
```
