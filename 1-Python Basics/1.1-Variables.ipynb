### Variables
Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In Python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a variable.

```python
a=100
```

```python
## Declaring and Assigning Variables

age = 25  # Assigning an integer value to a variable
name = "Alice"  # Assigning a string value to a variable
height = 5.7  # Assigning a float value to a variable
is_student = True  # Assigning a boolean value to a variable

## Printing Variable Values
print("Age:", age)
print("Name:", name)
print("Height:", height)
```

```python
## Naming Conventions
## Variable names should be descriptive and follow certain conventions:
# 1. Use lowercase letters and separate words with underscores (snake_case).
# 2. Avoid using reserved keywords (like if, else, while, etc.).
# 3. Start variable names with a letter or an underscore, not a number.

# valid variable names
# first_name = "John"
# last_name = "Doe" 
```

```python
# Invalid variable names
# 1name = "Invalid"  # Starts with a number
# class = "Invalid"  # Uses a reserved keyword
# last-name = "Invalid"  # Contains a hyphen
##@name="Victor"  # Contains a special character
```

```python
##case sensitivity
Var = 10
var = 20
Var==var  # This will be False because variable names are case-sensitive
```

```python
## Understanding Variable Types
## Python is a dynamically typed language, which means you don't need to declare the type of a variable explicitly. The type is inferred from the value assigned to it.
type(age)        # int
type(name)      # str
type(height)    # float
type(is_student) # bool

print("Type of age:", type(age))
print("Type of name:", type(name))
print("Type of height:", type(height))
print("Type of is_student:", type(is_student))
```

```python
age=30
print(type(age))

#Type Conversion
# You can convert between different variable types using built-in functions.
age_str = str(age)  # Convert integer to string
height_int = int(height)  # Convert float to integer
print("Age as string:", age_str)
print("Height as integer:", height_int)
print("Type of age_str:", type(age_str))
print("Type of height_int:", type(height_int))
```

```python
## Dynamic Typing
# You can reassign variables to different types.    
var = 10  # Initially an integer
print(var, type(var))

var = "Now I'm a string"  # Now reassigned to a string
print(var, type(var))

var = 3.14  # Now reassigned to a float
print(var, type(var))
```

```python
##Simple Calculator
# A simple calculator that takes two numbers and an operator from the user and performs the calculation.

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operator = input("Enter operator (+, -, *, /): ")
if operator == '+':
    result = num1 + num2
elif operator == '-':
    result = num1 - num2
elif operator == '*':
    result = num1 * num2    
elif operator == '/':
    result = num1 / num2
else:
    result = "Invalid operator"
print("Result:", result)

```

```python
sum=num1 + num2
difference=num1 - num2
product=num1 * num2
quotient=num1 / num2

print("Sum:", sum)
print("Difference:", difference)
print("Product:", product)
print("Quotient:", quotient)
```
