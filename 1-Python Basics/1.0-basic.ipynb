```python
## Basic Syntax Rules in Python
## Case Sensitivity
# Python is case-sensitive, meaning that variable names, function names, and other identifiers must be used consistently with regard to capitalization.

name="John"
Name="Doe"
print(name)  # Outputs: John
print(Name)  # Outputs: Doe
```

### Indentation
Indentation in Python is used to define the structure and heirarchy of the code. Unlike many other programming languages that use braces {} to delimit blocks of code, Python uses indentation to determine the grouping of statements. This ,eans that all the statements within a block must be indented at the same level.

```python
##Indentation
# Python uses indentation to define blocks of code. Consistent use of spaces or tabs is crucial for the correct execution of the code.

age = 18
if age >= 18:
    print(age)
```

```python
## This is a single line comment
print("Hello, World!")  # This is an inline comment
```

```python
## Line Continuation
# In Python, you can use a backslash (\) to indicate that a statement continues on the next line.

total = 1 + 2 + 3 +\
        4 + 5 + 6
print(total)  # Outputs: 21
```

```python
## Multiple statements on a Single Line
# You can write multiple statements on a single line using a semicolon (;).
x = 5; y = 10; print(x + y)  # Outputs: 15
```

```python
## Understanding Semantics in Python
# Semantics refers to the meaning of the code and how it behaves when executed. In Python, semantics are defined by the rules and conventions that govern how different constructs and statements operate.
#varable assignment
x = 10  # Assigns the integer value 10 to the variable x
y = "Hello"  # Assigns the string value "Hello" to the variable y
print(x)  # Outputs: 10
print(y)  # Outputs: Hello
```

```python
type(x)  # Outputs: <class 'int'>
type(y)  # Outputs: <class 'str'>
```

```python
## Type Inference
variable1 = 42          # Inferred as int
print(type(variable1))  # Outputs: <class 'int'>
variable2 = 3.14        # Inferred as float
print(type(variable2))  # Outputs: <class 'float'>
variable3 = "Python"    # Inferred as str
print(type(variable3))  # Outputs: <class 'str'> 
```
