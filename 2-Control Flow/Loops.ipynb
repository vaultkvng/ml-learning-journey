```python
## for loops in Python
ranges = range(5)
for i in ranges:
    print(i)
```

```python
## string iteration with for loop
message = "Hello"
for i in message:
    print(i)
```

```python
## while loops in Python
## the loop continues as long as the condition is True
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
count=0
while count%2 == 0:
    print(count)
    count += 1
```

```python
## loop control statements: break, continue, pass

## break statement
## the break statement terminates the loop entirely

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
## continue statement
## the continue statement skips the current iteration and moves to the next one
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
## pass statement
## the pass statement does nothing and is used as a placeholder
for i in range(5):
    if i == 3:
        pass  # Placeholder for future code
    print(i)
```

```python
## Nested loops
## A loop inside another loop
for i in range(3):
    for j in range(2):
        print(f"i: {i}, j: {j}")
```

```python
## Example 1: Sum of n natural numbers using for and while loops
n = 10
sum_for = 0
for i in range(1, n + 1):
    sum_for += i
print(f"Sum using for loop: {sum_for}")

sum_while = 0
count = 1
while count <= n:
    sum_while += count
    count += 1
print(f"Sum using while loop: {sum_while}")
```
