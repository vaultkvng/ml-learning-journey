```python
## 4.Write a program that asks the user to input a number and prints whether the number is positive and even, positive and odd or negative.
number = int(input("Please enter a number: "))
if number >= 0:
    if number % 2 == 0:
        print("The number is positive and even.")
    else:
        print("The number is positive and odd.")
else:
    print("The number is negative.")
    
```

```python
## 5. Write a program that prints all the numbers from 1 to 10 using a for loop.
for i in range(1, 11):
    print(i)
```

```python
##6. Write a program that prints all the numbers from 1 to 10 using a while loop.
i = 1
while i <= 10:
    print(i)
    i += 1
```

```python
## 7. Write a program that prints a 5x5 grid of asterisks (*) using nested loops.
for i in range(5):
    for j in range(5):
        print("*", end=" ")
    print()
```

```python
## 8. Write a program that asks the user to input numbers until they input 0. The program should print the sum of all the input numbers.
total = 0
while True:
    number = float(input("Please enter a number (0 to stop): "))
    if number == 0:
        break
    total += number
    print("The sum of all your inputed numbers is:", total)
```

```python
## 9. Write a program that prints all the numbers from 1 to 10, but skips the number 5 using a continue statement.
i = 1
while i <= 10:
    if i == 5:
        i += 1
        continue
    print(i)
    i += 1
```

```python
## 11. Write a program that asks the user to input a number and prints all the even numbers from 1 to that number using a for loop.
number = int(input("Please enter a number: "))
for i in range(1, number + 1):
    if i%2 == 0:
        print(i)
```

```python
##12. Write a program that calculates the factorial of a number input by the user using a while loop.
number = int(input("Please enter a number: "))
factorial = 1
i = 1
while i <= number:
    factorial *= i
    i += 1
print("The factorial of", number, "is", factorial)
```

```python
## 13. Write a program that calculates the sum of the digits of a number input by the user using a while loop.
number = int(input("Please enter a number: "))
sum_of_digits = 0
while number > 0:
    digit = number % 10
    sum_of_digits += digit
    number //= 10
print("The sum of the digits is:", sum_of_digits)
```

```python
## 14. Write a program that asks the user to input a number and checks whether it is a prime number.
number = int(input("Please enter a number: "))
prime_check = True
for i in range(2, int(number**0.5) + 1):
    if number % i == 0:
        prime_check = False
        break   
if prime_check and number > 1:
    print(number, "is a prime number.")
else:
    print(number, "is not a prime number.")
```

```python
##15. Write a program that prints the Fibonacci sequence up to a number input by the user.
number = int(input("Please enter a number: "))
a, b = 0, 1
while a <= number:
    print(a)
    a, b = b, a + b 
    
```
