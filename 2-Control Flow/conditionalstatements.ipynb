```python
## if statements in Python
age = 18
if age >= 18:
    print("You are eligible to vote.")
```

```python
age>=18
```

```python
## else statements in Python
## the else block executes when the if condition is False
age = 16
if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote yet.")
```

```python
## elif statements in Python
## the elif block allows checking multiple conditions
age = 17
if age < 13:
    print("You are a child.")
elif age < 20:
    print("You are a teenager.")
else:
    print("You are an adult.")
```

```python
## Nested if statements in Python
## You can nest if statements within other if statements
## Example: Check if a number is positive, negative, or zero and also check if it's even or odd
number=int(input("Enter a number: "))

if number > 0:
    print("The number is positive.")
    if number % 2 == 0:
        print("The number is also even.")
    else:
        print("The number is also odd.")
elif number < 0:
    print("The number is negative.")
else:
    print("The number is zero.")
    
```

```python
## Practical Examples
## Leap year checker
year = int(input("Enter a year: "))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(year, "is a leap year.")
        else:
            print(year, "is not a leap year.")
    else:
        print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

```python
## Determine ticket price based on age and student status
age = int(input("Enter your age: "))
is_student = input("Are you a student? (yes/no): ").lower()
if age < 12:
    print("Ticket price: $5")   
elif age <= 18:
    print("Ticket price: $7")
elif is_student == 'yes':
    print("Ticket price: $8")
else:
    print("Ticket price: $10")


```
